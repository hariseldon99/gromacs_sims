```python
import numpy as np
import matplotlib.pyplot as plt

# Load data from xvg file
data = np.loadtxt('distav.xvg', comments=['#', '@'])

# Extract columns from data
x = data[:, 0]/1000
y = data[:, 1]

# Plot data using Matplotlib
plt.plot(x, y, linewidth = 0.3)
plt.fill_between(x, y, color = 'cyan', alpha = 0.3)
plt.xlabel('Time (ns)')
plt.ylabel('Distance (nm)')
plt.title('Distance between atom id 1236 and 6599')
plt.grid(True, linestyle = '--')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick

from matplotlib import rc
plt.rcParams["font.family"] = "sans-serif"
plt.rcParams["font.sans-serif"] = ["Nimbus Sans"]


font = {'family' : 'sans-serif',
        'size'   : 12}

rc('font', **font)
ax = plt.gca()
ax.yaxis.set_major_formatter(mtick.PercentFormatter())

# Load data from xvg file
data = np.loadtxt('distav.xvg', comments=['#', '@'])

# Extract columns from data
x = data[:, 0]/1000
y1 = data[:, 1]

cutoff_dist = 0.35

#eta bhul cilo
#y = ((y1-cutoff_dist)/y1)*100
y = ((y1-cutoff_dist)/cutoff_dist)*100

pos_x = x.copy()
pos_y = y.copy()

pos_x[pos_y<=0] = np.nan
pos_y[pos_y<=0] = np.nan


neg_x = x.copy()
neg_y = y.copy()

neg_x[neg_y>=0] = np.nan
neg_y[neg_y>=0] = np.nan

# Plot data using Matplotlib
ax.plot(pos_x, pos_y,color='red')
ax.plot(neg_x, neg_y,color='blue')

ax.set_xlabel('Time (ns)')
ax.set_ylabel('% Deviation')
ax.set_title('Deviation from cutoff: Atom id 1236 (MMP2) \& 6599 (Rol) ')
ax.axhline(0, color = 'red', linestyle = '--')
ax.grid(True, linestyle = '--')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rc
import matplotlib.ticker as mtick

plt.rcParams["font.family"] = "sans-serif"
plt.rcParams["font.sans-serif"] = ["Nimbus Sans"]


font = {'family' : 'sans-serif',
        'size'   : 12}

rc('font', **font)
ax = plt.gca()
ax.yaxis.set_major_formatter(mtick.PercentFormatter())

# Load data from xvg file
data = np.loadtxt('angaver.xvg', comments=['#', '@'])

# Extract columns from data
x = data[:, 0]/1000
y1 = data[:, 1]

cutoff_angle = 150
y = ((cutoff_angle - y1)/cutoff_angle)*100

pos_x = x.copy()
pos_y = y.copy()

pos_x[pos_y<=0] = np.nan
pos_y[pos_y<=0] = np.nan


neg_x = x.copy()
neg_y = y.copy()

neg_x[neg_y>=0] = np.nan
neg_y[neg_y>=0] = np.nan


# Plot data using Matplotlib
ax.plot(pos_x, pos_y,color='red')
ax.plot(neg_x, neg_y,color='blue')

ax.set_xlabel('Time (ns)')
ax.set_ylabel('% Deviation')
plt.title('Ang. Devn. from cutoff: Atom ids 1236, 1237 (MMP2) \& 6599 (Rol) ')
ax.axhline(0, color = 'red', linestyle = '--')
ax.grid(True, linestyle = '--')
plt.show()
```
